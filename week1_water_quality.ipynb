```python
# Week 1 - Water Quality Prediction (AI + Green Technology)

# ---------------------------
# Step 1: Import Libraries
# ---------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------
# Step 2: Load Dataset
# ---------------------------
# Upload your Kaggle CSV "water_potability.csv" in Colab first (use left sidebar > Files > Upload)
data = pd.read_csv("water_potability.csv")

print("Dataset Shape:", data.shape)
print("\nFirst 5 rows of dataset:")
print(data.head())

# ---------------------------
# Step 3: Understand Data
# ---------------------------
print("\nDataset Info:")
print(data.info())

print("\nSummary Statistics:")
print(data.describe())

print("\nMissing Values in Each Column:")
print(data.isnull().sum())

# ---------------------------
# Step 4: Handle Missing Values (Preprocessing)
# ---------------------------
# Filling missing values with median (robust for skewed data)
data = data.fillna(data.median())

print("\nMissing Values After Filling:")
print(data.isnull().sum())

# ---------------------------
# Step 5: Data Visualization
# ---------------------------

# Distribution of Target (Potability)
plt.figure(figsize=(6,4))
sns.countplot(x="Potability", data=data, palette="Set2")
plt.title("Distribution of Safe vs Unsafe Water Samples")
plt.xlabel("Potability (0 = Unsafe, 1 = Safe)")
plt.ylabel("Count")
plt.show()

# Histogram of Features
data.hist(figsize=(12,10), bins=30, color="skyblue", edgecolor="black")
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()

# Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(), annot=False, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()

# ---------------------------
# Step 6: Problem Statement
# ---------------------------
problem_statement = """
Problem Statement:
------------------
We aim to predict whether water is potable (safe for drinking) based on chemical properties
such as pH, hardness, solids, chloramines, sulfate, conductivity, organic carbon,
trihalomethanes, turbidity, etc.

Why It Matters:
---------------
Unsafe water causes millions of deaths annually.
Using Machine Learning, we can help authorities rapidly screen water samples
without waiting for costly lab results.

This aligns with UN Sustainable Development Goal (SDG) 6:
'Ensure availability and sustainable management of water and sanitation for all.'
"""
print(problem_statement)
```
